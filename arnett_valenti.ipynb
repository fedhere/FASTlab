```python
DAYS2SEC=24.0*3600.0
SEC2DAYS=1.0/3600./24.
Msun2g=2.0e33
kopt=0.06 #cm*cm/g
beta=13.8
c=3.0e10#constants.c*100 #cm/s

x=np.arange(0.01,600,1.0)
M=np.arange(0.1,10,1.0)
v=np.arange(5,30,1.0)
v=v*1e4*1e3*100.0 #cm/s
y=[np.sqrt(2.0*kopt/beta/c * M*1e33/vi)/DAYS2SEC for vi in v]
y= np.array(y).flatten()
#print y
eni=3.9e10 #erg/s/g
eco=6.78e9 #erg/s/g
tni=6.*DAYS2SEC #days,s
tco=77.*DAYS2SEC  #days,s
```

```python
def s(tm):
    return tm*(tco-tni)/(2.0*tco*tni)

def intA(z,tm):
#    x=t/tm
    y=tm/(2.0*tni)
    return 2.0*z*np.exp(-2*z*y+z*z)

def intB(z,tm):
#    x=t/tm
    y=tm/(2.0*tni)
    return 2.0*z*np.exp(-2*z*y+z*z+2.0*z*s(tm))

def valentilum(Mni,tm,x):
    return (Mni*Msun2g)* np.exp(-(x*x)) * ( (eni-eco)*quad(intA,0,x,args=(tm))[0] + eco*quad(intB,0,x,args=(tm))[0] )

integrator=np.linspace(0,1,1000.,endpoint=False)
def valentilum2(Mni,tm,x):
    xs=integrator*x
    dz=xs[1]-xs[0]
    return (Mni*Msun2g)* np.exp(-(x*x)) * ( (eni-eco)*intA(xs,tm) + eco*intB(xs,tm)).sum()*dz
```

```python
def lnlike(theta,ypar,yerr,plot=False,sigs=False):
    """
    minus log likelihood , y = mag, dm15, v
    """
    M,Mni,E=theta
    
    model=np.array(arnett(M,Mni,E))
    
    #print "ypar",ypar, model, yerr
    #print "ypar-model",ypar-model
    #print "ypar",model,ypar
    return -0.5*(np.sum((ypar-model)**2/yerr**2 ))

def lnprior(theta):
    """PRIOR:
    M>0.1
    M<10
    Mni>0.01 Msun
    Mni<5 Msun
    E>0.1<100 (e51)

    """   
    M,Mni,E=theta
    if 0.1<M<10 and 0.01<Mni<3 and 0.1<E<100:
        return 0.0
    return -np.inf

def lnprob(theta, ypar, yerr):
    lp = lnprior(theta)
    if not (np.isfinite(lp)):
        return -np.inf
    thislnlike=lnlike(theta, ypar, yerr)
    if np.isnan(thislnlike):
        return -np.inf
    #print lp + thislnlike
    return lp + thislnlike
```

```python

LSunbol= 3.846e33#erg/s
MSunbol=4.74
def getbolmag(Lbol):
    LSunbol= 3.846e33#erg/s
    MSunbol=4.74
    #     print Lbol
    #     print  MSunbol-2.5*np.log10(np.array(Lbol)/LSunbol)
    return MSunbol-2.5*np.log10(Lbol/LSunbol)
```

```python
#diffusion time scale 
tmconst=  np.sqrt(kopt/beta/c) * ((Msun2g)**(3.0/4.0)) * ((1.0e51)**-0.25) * (10.0/3.0)**(3.0/4.0) #*1000
##[g^-0.5 cm^0.5 s^0.5 * g^0.25 *g^0.5  g^-0.25 cm^-0.5 *s^0.5]
##[ s^0.5  * s^0.5] = [s]
t=np.arange(0.5,61,0.1)*DAYS2SEC
```

```python
def arnett_lum(M,Mni,E):
    tm=pow( M**3/E, 0.25)
    tm*=tmconst*0.33
    y=tm/(2.0*tni)
    x=t/tm 
    lum=np.array([ valentilum2(Mni,tm,thisx) for thisx in x])
    lum[~np.isfinite(lum)]=float('NaN')
    return lum
```

```python
def arnett(M,Mni,E, plot=False):
    #print "pars",M,Mni,E
    if M>15 or M<0.01 or Mni>10 or Mni<0.001 or E>100 or E<0.05:
        return np.array([float('NaN'),float('NaN'),float('NaN')])
    #print "truth ratio", M/3.,Mni/0.36,E/7.
    
    lum=arnett_lum(M,Mni,E)
    if plot:
        pl.plot(t*SEC2DAYS,lum,'k-')
        
    v=np.sqrt(10.0/3.0*E*1.0e51/M/2.0e33)*1e-5
    #print "here",-getbolmag(lum[np.nanargmax(lum)])+getbolmag(lum[np.nanargmax(lum)+150]),(lum[np.nanargmax(lum)])*1.0e-42,v 
    maxplace=np.nanargmax(lum)
    if maxplace+150>len(lum)-1:
        return np.array([float('NaN'),float('NaN'),float('NaN')])
        
    #    return np.nanargmax(lum)*10,(getbolmag(lum[np.nanargmax(lum)])-getbolmag(lum[np.nanargmax(lum)+150]),getbolmag(lum[np.nanargmax(lum)]),v,np.log10(lum[np.argmax(lum)]),np.log10(lum[np.argmax(lum)+150]))
    return [2.5*np.log10(lum[maxplace]/lum[maxplace+150]),lum[maxplace]*1.0e-42,v*1e-4]
```
